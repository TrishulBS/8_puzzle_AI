```python
import heapq
import matplotlib.pyplot as plt
import datetime
import time


def main():
    
    #Prompt to accept user input whether to solve test case puzzle or own puzzle
    user_input= input("welcome to 8-puzzle solver. If you want to solve test case puzzle, then press 't'. If you want to solve your arbitrary puzzle then press 'a'")

    #Continue the loop until user enters valid option
    while user_input not in ['t', 'a']:
        print("you have entered wrong input. Please enter 't' or 'a'")
        user_input= input("welcome to 8-puzzle solver. If you want to solve test case puzzle, then press 't'. If you want to solve your arbitrary puzzle then press 'a'")

    if user_input == 't':
        
        #test cases
        default_puzzles = {
            1: [[1,2,3], [4,5,6], [7,8,0]],
            2: [[1,2,3], [4,5,6], [0,7,8]],
            3: [[1,2,3], [5,0,6], [4,7,8]],
            4: [[1,3,6], [5,0,2], [4,7,8]],
            5: [[1,3,6], [5,0,7], [4,8,2]],
            6: [[1,6,7], [5,0,3], [4,8,2]],
            7: [[7,1,2], [4,8,5], [6,3,0]],
            8: [[0,7,2], [4,6,1], [3,5,8]]
        }

        #Loop until user enters valid test case puzzle number
        while True:
            #prompt to enter test case puzzle number
            print('Which puzzle do you want to solve? Press the corresponding number')
            for key in default_puzzles:
                print(key, default_puzzles[key])

            
            user_input = int(input())

            if user_input in default_puzzles:
                initial_puzzle = default_puzzles[user_input]
                algorithm_selection(initial_puzzle, [[1,2,3], [4,5,6], [7,8,0]])
                break
            else:
                print('You entered an invalid option. Please try again!')
    else:
        #Solve arbitrary puzzle
        own_puzzle = []
        print('Be kind to the system when giving the input')
        print('what should be the length of the puzzle')
        
        #User input of size of the puzzle
        length = int(input())



        own_puzzle_dest = []
        counter = 1
        
        #Filling the goal state of the arbitrary puzzle
        for k in range(length):
            individual_row = []
            for m in range(length):
                if k == length - 1 and m == length - 1:
                    individual_row.append(0)
                else:
                    individual_row.append(counter)
                    counter += 1
            own_puzzle_dest.append(individual_row)


        # Accepting initial state of the arbitrary puzzle
        while True:
            print(f"Enter each individual rows of length {length}")
            for i in range(1, length + 1):
                user_input = list(map(int, input().split()))
                own_puzzle.append(user_input)

            #Check if the arbitrary puzzle has any duplicate values
            flattened_list = [num for sublist in own_puzzle for num in sublist]
            print(own_puzzle, flattened_list)
            if len(flattened_list) != len(set(flattened_list)):
                print("The list of lists contains duplicate numbers.")
                own_puzzle = []
                continue
            else:
                break

        # Algorithm selection and execution
        algorithm_selection(own_puzzle, own_puzzle_dest)


if __name__ == '__main__':
    main()
```

```python
def algorithm_selection(start_state, final_state):

    # Prompt user for which algorithm to execute
    while True:
        print('What search algorithm do you want to use')
        print('1. Unform Cost Search')
        print('2. A* with Manhattan distnace Heuristic')
        print('3. A* with Misplaced Tile Heuristic')

        user_input = int(input())

        if user_input in [1,2,3]:
            #Algorithm exceution time calculation
            current1 = time.time()
            #Execution of the chosen algorithm
            algorithm_execution(start_state, final_state, user_input)
            current2 = time.time()
            time_taken = int(round((current2 - current1) * 1000))
            print(f"Algorithm run duration: {time_taken} milliseconds")
            break  
        else:
            print('You entered an invalid input. Please try again')
```

```python
def algorithm_execution(initial_st, end_st, algorithm):
    

    search_iterations = 0 
    search_complexity = 1
    #To store states that are already explored
    explored = []
    #to store states in the queue to be explored
    pending = []
    #Create tree strctuture for the states
    root = Create_Tree(initial_st)
    #Usage of priororty queue since min heuristics need to be popped
    heapq.heappush(pending, root)

    #Check if he queue is empty
    while pending:

        if len(pending) > search_complexity :
            search_complexity = len(pending)

        #pop the state which has minimum heuristic from the queue
        present = heapq.heappop(pending)

        present_state = present.state
        
        boolean=True
#         if len(present_state) != len(end_st) or len(present_state[0]) != len(end_st[0]):
#             boolean=False
#         else:

        # Check if the curent state is te goal state
        for r in range(len(present_state)):
            for c in range(len(present_state[0])):
                if present_state[r][c] != end_st[r][c]:
                    boolean=False
                    break
            if boolean == False:
                break
                
        #If current state is the goal state, then return 
        if boolean == True:
            print('Puzzle Solved')
            print('The path on the tree from root to goal are as follows')
            goal_to_init_path = []
            pointer = present
            #Print all the nodes along the path from initial state to goal state
            while(pointer!=None):
                goal_to_init_path.append(pointer.state)
                pointer = pointer.parent
            init_to_goal_path = goal_to_init_path[::-1]
            count=0
            for x in init_to_goal_path:
                print(x)
                count+=1
            print('Number of nodes from initial state to final state is ', count)
            print('Maximum number of nodes in the queue is', search_complexity)
            print('Number of nodes expanded is', search_iterations)
            return
        
        else:
            #If current state is not a goal state
            explored.append(present)
            # Calculate all possible states from the current state
            next_moves = present.next_states()
            
            
            #For each next move(state), calculate Heuristic based on the algorithm and add 
            #to prioritu queue
            for next_move in next_moves:
                if next_move == None:
                    continue
                #Create node for the next states
                next_move_node = Create_Tree(next_move)
                
                #Only add to the prioriy queue if it is not explored or in the queue
                if next_move_node not in explored and next_move_node not in pending:
                    if algorithm == 1:
                        #heuristic is zero for Uniform cost search
                        next_move_node.heuristic = 0
                    elif algorithm == 2:
                        #calculation of manhattan heuristics
                        heuristic_manhattan = 0
                        start_pos = {}
                        goal_pos = {}
                        for k in range(len(end_st)):
                            for m in range(len(end_st)):
                                start_pos[next_move_node.state[k][m]] = (k, m)
                                goal_pos[end_st[k][m]] = (k, m)

                        for num in range(1, len(end_st) ** 2):
                            start_row, start_col = start_pos[num]
                            goal_row, goal_col = goal_pos[num]
                            #Calculation of manhattan distance from its goal state to current state
                            heuristic_manhattan += abs(start_row - goal_row) + abs(start_col - goal_col)
                        next_move_node.heuristic = heuristic_manhattan
                    else:
                        #Calculation of misplace tile heuristics
                        heuristic_misplaced = sum(1 for k in range(len(next_move_node.state)) for m in range(len(next_move_node.state))
                          if next_move_node.state[k][m] != end_st[k][m] and next_move_node.state[k][m] != 0)
                        next_move_node.heuristic = heuristic_misplaced
                  
                #Addth next states from the current state as children to the cuurent state
                present.addchild(child_node = next_move_node)
                #Push the next states to be explored to the priority queue
                heapq.heappush(pending, next_move_node)
            
            search_iterations += 1
    
    #If search did not find solution, then return with no solution
    if len(pending) == 0:
        print("Cannot find the solution")
        return
```

```python
import copy
#Create Tree structue to explore search space
class Create_Tree:
    def __init__(self, state=0, heuristic=0, path_cost=0, parent=None):
        #State of the node
        self.state = state
        # heuristic value of the node based on algorithm, default is zero
        self.heuristic = heuristic
        #path cost of the node
        self.path_cost = path_cost
        #Store parent state from this state
        self.parent = parent
        #Store children from this state
        self.children = []
        
    # method to keep track of parent
    def addchild(self, child_node, transition_cost=1):
        child_node.path_cost = self.path_cost + transition_cost
        child_node.parent = self
        self.children.append(child_node)
        

    #Used for peiority queue operation for checking the minimum heuristic
    def __lt__(self, other):
        node_sum = self.path_cost + self.heuristic
        other_sum = other.path_cost + other.heuristic
        return node_sum < other_sum
    
    
    #Generate al possible next states from current state
    def next_states(self):
        rw, cl = 0, 0
        next_sts = []
        puzzle = self.state
        

       
        found_zero = False
        for k, row in enumerate(puzzle):
            if found_zero:
                break
            for m, value in enumerate(row):
                if value == 0:
                    rw, cl = k, m
                    found_zero = True
                    break

    
        directions = [(0, -1, 'left'), (-1, 0, 'up'), (0, 1, 'right'), (1, 0, 'down')]

        for direction in directions:
            new_m, new_n, direction_name = rw + direction[0], cl + direction[1], direction[2]

            if (0 <= new_m < len(puzzle)) and (0 <= new_n < len(puzzle)):
                newstate = copy.deepcopy(puzzle)
                newstate[rw][cl], newstate[new_m][new_n] = newstate[new_m][new_n], newstate[rw][cl]
                
                boolean = True
                if newstate == None or self.parent == None:
                    boolean = False
                elif len(newstate) != len(self.parent.state) or len(newstate[0]) != len(self.parent.state[0]):
                    boolean = False
                else: 
                    for k in range(len(newstate)):
                        for m in range(len(newstate[0])):
                            if newstate[k][m] != self.parent.state[k][m]:
                                boolean = False
                                break
                        if boolean == False:
                            break

                if (self.parent and boolean):
                    next_sts.append(None)
                else:
                    next_sts.append(newstate)

        return next_sts
```
